## HAL Summary Plot Generator

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
right_stringer_data = pd.read_excel(r"C:\Users\anand\Desktop\HAL Predictor Summary Statistics\right_stringer_data.xlsx")
right_camel_hump_data = pd.read_excel(r"C:\Users\anand\Desktop\HAL Predictor Summary Statistics\right_camel_hump_data.xlsx")
left_stringer_data = pd.read_excel(r"C:\Users\anand\Desktop\HAL Predictor Summary Statistics\left_stringer_data.xlsx")
left_camel_hump_data = pd.read_excel(r"C:\Users\anand\Desktop\HAL Predictor Summary Statistics\left_camel_hump_data.xlsx")

# Column names are
#Participant ID,Accuracy, True Low Predicted Medium, True Low Predicted High,True Medium Predicted High, True Medium Predicted Low,
#True High Predicted Medium, True High Predicted Low
```

### Creating figures for the paper

```python
# 4 subplots in the same figure, one for each data set and x-axis label only on the bottom
bar_width = 0.3
indices = np.arange(15)

# Right Hand, Stringer Tool
correct_or_conservative = right_stringer_data['Accuracy']+right_stringer_data['True Low Predicted Medium']+right_stringer_data['True Medium Predicted High']
incorrect_by_one = right_stringer_data['True Medium Predicted Low']+right_stringer_data['True High Predicted Medium']
incorrect_by_two = right_stringer_data['True High Predicted Low']+right_stringer_data['True Low Predicted High']
data = pd.DataFrame({'Correct or Conservative': correct_or_conservative, 'Incorrect by One': incorrect_by_one, 'Incorrect by Two': incorrect_by_two})
fig, axs = plt.subplots(4,1, figsize=(14, 17), sharex=True)
axs[0].bar(indices - bar_width, data['Correct or Conservative'], width=bar_width, color='green',alpha = 0.6, label='Correct or Conservative')
axs[0].bar(indices, data['Incorrect by One'], width=bar_width, color='#d4aa00', alpha = 0.5, label='Incorrect by One')  # Darker yellow
axs[0].bar(indices + bar_width, data['Incorrect by Two'], width=bar_width, color='#ff9999', label='Incorrect by Two')  # Lighter red
# Horizontal line at y = 0.95 with the text '95% accuracy' at the left end
axs[0].axhline(y=0.95, color='black', linestyle='--')
axs[0].text(0.2, 0.97, '95% accuracy', va='center', ha='right')
axs[0].set_ylabel('Fraction of Data')
axs[0].set_title('Right Hand, Stringer Tool')
axs[0].legend(bbox_to_anchor=(0.001, 1.3), loc='upper left')
# axs[0].legend()

correct_or_conservative = right_camel_hump_data['Accuracy']+right_camel_hump_data['True Low Predicted Medium']+right_camel_hump_data['True Medium Predicted High']
incorrect_by_one = right_camel_hump_data['True Medium Predicted Low']+right_camel_hump_data['True High Predicted Medium']
incorrect_by_two = right_camel_hump_data['True High Predicted Low']+right_camel_hump_data['True Low Predicted High']
data = pd.DataFrame({'Correct or Conservative': correct_or_conservative, 'Incorrect by One': incorrect_by_one, 'Incorrect by Two': incorrect_by_two})

axs[1].bar(indices - bar_width, data['Correct or Conservative'], width=bar_width, color='green',alpha = 0.6, label='Correct or Conservative')
axs[1].bar(indices, data['Incorrect by One'], width=bar_width, color='#d4aa00', alpha = 0.5, label='Incorrect by One')  # Darker yellow
axs[1].bar(indices + bar_width, data['Incorrect by Two'], width=bar_width, color='#ff9999', label='Incorrect by Two')  # Lighter red
axs[1].axhline(y=0.95, color='black', linestyle='--')
axs[1].text(0.2, 0.97, '95% accuracy', va='center', ha='right')
axs[1].set_ylabel('Fraction of Data')
axs[1].set_title('Right Hand, Camel Hump Tool')
# axs[1].legend()

correct_or_conservative = left_stringer_data['Accuracy']+left_stringer_data['True Low Predicted Medium']+left_stringer_data['True Medium Predicted High']
incorrect_by_one = left_stringer_data['True Medium Predicted Low']+left_stringer_data['True High Predicted Medium']
incorrect_by_two = left_stringer_data['True High Predicted Low']+left_stringer_data['True Low Predicted High']

data = pd.DataFrame({'Correct or Conservative': correct_or_conservative, 'Incorrect by One': incorrect_by_one, 'Incorrect by Two': incorrect_by_two})

axs[2].bar(indices - bar_width, data['Correct or Conservative'], width=bar_width, color='green',alpha = 0.6, label='Correct or Conservative')
axs[2].bar(indices, data['Incorrect by One'], width=bar_width, color='#d4aa00', alpha = 0.5, label='Incorrect by One')  # Darker yellow
axs[2].axhline(y=0.95, color='black', linestyle='--')
axs[2].text(0.2, 0.97, '95% accuracy', va='center', ha='right')
axs[2].bar(indices + bar_width, data['Incorrect by Two'], width=bar_width, color='#ff9999', label='Incorrect by Two')  # Lighter red
axs[2].set_ylabel('Fraction of Data')
axs[2].set_title('Left Hand, Stringer Tool')

correct_or_conservative = left_camel_hump_data['Accuracy']+left_camel_hump_data['True Low Predicted Medium']+left_camel_hump_data['True Medium Predicted High']
incorrect_by_one = left_camel_hump_data['True Medium Predicted Low']+left_camel_hump_data['True High Predicted Medium']
incorrect_by_two = left_camel_hump_data['True High Predicted Low']+left_camel_hump_data['True Low Predicted High']
data = pd.DataFrame({'Correct or Conservative': correct_or_conservative, 'Incorrect by One': incorrect_by_one, 'Incorrect by Two': incorrect_by_two})

axs[3].bar(indices - bar_width, data['Correct or Conservative'], width=bar_width, color='green',alpha = 0.6, label='Correct or Conservative')
axs[3].bar(indices, data['Incorrect by One'], width=bar_width, color='#d4aa00', alpha = 0.5, label='Incorrect by One')  # Darker yellow
axs[3].axhline(y=0.95, color='black', linestyle='--')
axs[3].text(0.2, 0.97, '95% accuracy', va='center', ha='right')
axs[3].bar(indices + bar_width, data['Incorrect by Two'], width=bar_width, color='#ff9999', label='Incorrect by Two')  # Lighter red
axs[3].set_title('Left Hand, Camel Hump Tool')

plt.xlabel('Participant ID')
plt.ylabel('Fraction of Data')

plt.xticks(indices, [f'Participant {i+1}' for i in range(15)], rotation=45)
plt.suptitle('Summary Statistics of Gated Recurrent Unit Predictor for HAL Score', fontsize=14)
plt.tight_layout(pad=1.0)
plt.savefig(r'C:\Users\anand\Desktop\ergo paper figures\HAL Predictor Summary Statistics.png')
```
